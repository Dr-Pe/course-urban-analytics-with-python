# Overpass API
### Acquiring point data

## First look at the API: City of Buenos Aires

```python
import overpy  # https://python-overpy.readthedocs.io/en/latest/api.html
import geopandas as gpd
import matplotlib.pyplot as plt

from shapely import Point
```

```python
api = overpy.Overpass(url="https://overpass.private.coffee/api/interpreter")
```

```python
miny = -34.71
minx = -58.53
maxy = -34.53
maxx = -58.34

amenity_cat = "amenity"
amenity_name = "cafe"
bbox = ",".join([str(miny), str(minx), str(maxy), str(maxx)])

query = f"""node["{amenity_cat}"="{amenity_name}"]({bbox});out;"""

query
```

```python
result = api.query(query)
```

```python
poi_data = []

for node in result.nodes:
    poi_data.append({
            "id": node.id,
            "name": node.tags.get("name", "N/A"),
            "amenity": node.tags.get("amenity"),
            "geometry": Point(node.lon, node.lat),
    })

gdf = gpd.GeoDataFrame(poi_data)

gdf.head()
```

```python
gdf.crs = "EPSG:4326"
gdf.explore(marker_type="circle", marker_kwds={"radius": 50, "fill": True})
```

```python
gdf.to_csv("data/raw/cafes_de_buenos_aires.csv")
```
