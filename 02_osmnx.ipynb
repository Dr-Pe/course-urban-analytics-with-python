# OSMnx

```python
import osmnx as ox
import matplotlib.pyplot as plt
import pandas as pd
```

## Acquiring administrative areas and extending visuality

```python
caba_gdf = ox.geocode_to_gdf("CABA")
com_1_gdf = ox.geocode_to_gdf("R4261029", by_osmid=True)
```

```python
fig = pd.concat([caba_gdf, com_1_gdf]).plot(
    "name", legend=True, edgecolor="black", linewidth=0.8
)

fig.get_legend().set_bbox_to_anchor((1.5, 1))
fig.get_legend().set_frame_on(False)
fig.set_frame_on(False)
```

## Acquiring polygon-level urban features
### Parks & buildings

```python
parks = ox.features_from_polygon(com_1_gdf.geometry.iloc[0], tags={"leisure": "park"})
buildings = ox.features_from_polygon(
    com_1_gdf.geometry.iloc[0], tags={"building": True}
)
```

```python
f, ax = plt.subplots()

com_1_gdf.plot(ax=ax, color="none", edgecolor="black", linewidth=0.8)
parks.plot(ax=ax, color="green")
buildings.plot(ax=ax, color="grey")

plt.show()
```

## Acquiring graph data

```python
G = ox.graph_from_polygon(com_1_gdf.geometry.iloc[0])
```

```python
nodes, edges = ox.graph_to_gdfs(G)
```

```python
f, ax = plt.subplots()

com_1_gdf.plot(ax=ax, color="none", edgecolor="black")
nodes.plot(ax=ax, color="black", markersize=1)
edges.plot(ax=ax, color="grey", linewidth=0.5)

plt.show()
```
